```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Clean_Data/cleaned_data.csv")
df.columns = df.columns.str.strip()

gender_colors = {'male': 'rosybrown', 'female': 'navy'}


# Visualization 1: Box Plot - Math Scores by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math score', data=df, palette=gender_colors)
plt.title('Math Scores Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Clean_Data/cleaned_data.csv")

# Visualization 2: Bar Chart - Average Reading Scores by Parental Education Level
plt.figure(figsize=(10, 6))
edu_avg = df.groupby("parental level of education")[["reading score"]].mean().sort_values(by="reading score", ascending=False)
edu_avg.plot(kind='bar', legend=False, colormap="viridis")
plt.title("Average Reading Scores by Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Average Reading Score")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Clean_Data/cleaned_data.csv")

df.columns = df.columns.str.strip()
education_colors = {'high school': 'blue', 'some college': 'plum', "bachelor's degree":'lightslategray', "master's degree":'teal',
                     "associate's degree":'peru','some high school':'crimson'}

# Violin Plot - Distribution of Math Scores by Parental Education Level
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='parental level of education', y='math score', palette="coolwarm")
plt.xticks(rotation=45)
plt.title('Distribution of Math Scores by Parental Education Level')
plt.xlabel('Parental Level of Education')
plt.ylabel('Math Score')
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Clean_Data/cleaned_data.csv")
# Strip whitespaces from column names
df.columns = df.columns.str.strip()
# Pie Chart - Distribution of Lunch Status by Parental Education Level
plt.figure(figsize=(10, 8))
lunch_counts = df.groupby("parental level of education")["lunch"].value_counts().unstack()
lunch_counts.plot(kind="pie", subplots=True, autopct="%1.1f%%", startangle=140, figsize=(12, 6), legend=False)
plt.suptitle("Distribution of Lunch Status by Parental Education Level")
plt.ylabel("")  # Remove y-axis label for clarity
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Clean_Data/cleaned_data.csv")
# Strip whitespaces from column names
df.columns = df.columns.str.strip()
gender_colors = {'male': 'rosybrown', 'female': 'navy'}

# Swarm Plot - Math Score by Test Preparation Course and Gender
plt.figure(figsize=(8, 6))
sns.swarmplot(data=df, x='test preparation course', y='math score', hue='gender', palette=gender_colors, dodge=True)
plt.title('Math Score Distribution by Test Preparation Course and Gender')
plt.xlabel('Test Preparation Course')
plt.ylabel('Math Score')
plt.show()
```
